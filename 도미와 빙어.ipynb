```python
import numpy as np
fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

# fish_data = [[l, w] for l,w in zip(fish_length, fish_weight)]
# fish_target = [1]*35+[0]*14
fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data)
fish_target = np.concatenate((np.ones(35), np.zeros(14)))#np.ones(), np.zeros()는 각각 원하는 만큼의 1, 0 로 채워진 배열을 만들어준다
print(fish_target)
```

<h1>#훈련 데이터 셋이랑 테스트 데이터 셋 나누기</h1>

```python

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state = 42)#train_test_split을 이용하여 데이터 나누기
print(train_input.shape, train_target.shape)
print(test_input.shape, test_target.shape)
print(test_target) #test_target셋의 빙어의 비율이 모자르다
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify = fish_target, random_state = 42)#stratify 매개변수에 타깃 데이털르 전달하면 클래스 비율에 맞게 데이터를 나눈다
#훈련 데이터가 작거나 특정 클래스의 샘플 개수가 적을때 사용
print(test_target)

# train_input = fish_data[:35]

# train_target = fish_target[:35]

# test_input = fish_data[35:]

# test_target = fish_target[35:]



```

<h1>#데이터 섞기</h1>

```python
import numpy as np
#넘파이 사용해서 배열 생성

input_arr = np.array(fish_data)

target_arr = np.array(fish_target)

np.random.seed(42)
index = np.arange(49)
np.random.shuffle(index)

train_input = input_arr[index[:35]]
train_target = target_arr[index[:35]]


test_input = input_arr[index[35:]]
test_target = target_arr[index[35:]]
print(train_input)
print(train_target)
print(test_input)
print(test_target)

import matplotlib.pyplot as plt

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(test_input[:,0], test_input[:,1])
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

<h1>#knn 알고리즘을 이용한 머신러닝 프로그램 훈련시키기</h1>

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()

kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

<h1>#[25, 150]사이즈의 생선 종류 예측</h1>

```python
print(kn.predict([[25, 150]]))
```

<h1># [25, 150]사이즈의 생선이 왜 빙어로 판별될까?</h1>

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
distances, indexes = kn.kneighbors([[25, 150]]) #kneighbors() 괄호 안의 값과 가장 가까운 이웃까지의 거리와 인덱스 반환함. k값이 5명 가까운 5개까지 반환

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker = 'D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

print(indexes, distances)
print(train_input[indexes])
print(train_target[indexes])
```

<h1>#x축과 y축의 비율이 이상하단걸 파악->데이터의 스케일을 맞춰줘야함</h1>

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker = 'D')
plt.xlim((0, 1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
mean = np.mean(train_input, axis = 0) #np.mean() =>평균을 계산, axis = 0 이면 행을 따라 각 열의 통계값을 계산한다.
std = np.std(train_input, axis=0) #np.std()=>표준 편차를 계산, axis = 0 이면 행을 따라 각 열의 통계값을 계산한다.
print(mean, std)
```

<h1>#표준점수로 데이터 크기 맞추기</h1>

```python
train_scaled = (train_input-mean)/std
```

```python
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25, 150, marker = '^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

<h1>#문제의 그 데이터 값도 평균을 빼고 표준 편차로 나눠서 스케일링 해줘야함</h1>

```python
scaled_sample = ([25, 150]-mean)/std
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(scaled_sample[0], scaled_sample[1], marker = '^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

<h1>#스케일링된 데이터로 모델 훈련하기</h1>

```python
kn.fit(train_scaled, train_target)

test_scaled = (test_input-mean)/std

kn.score(test_scaled, test_target)

print(kn.predict([scaled_sample]))
```

<h1>#데이터 스케일링 후 kneighbors()메서드로 반환되는 값의 변화 확인하기</h1>

```python
distances, indexes = kn.kneighbors([scaled_sample])
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(scaled_sample[0], scaled_sample[1], marker='^')
plt.scatter(train_scaled[indexes, 0], train_scaled[indexes, 1], marker = 'D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
